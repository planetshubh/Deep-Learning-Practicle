```python
# python helper functions for plotting graphs -
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
import os

plt.style.use('fivethirtyeight')

x= np.linspace(-10, 10, 100)

root_plot_dir = "all plots"
os.makedirs(root_plot_dir, exist_ok = True)

def graph_plot(x,func,
               ALPHA = 0.6,
               label_x = r"$x \rightarrow$", label_y = r"$act(x)  /\ act'(x) \rightarrow$",
               title=None,
               LABEL_y=None,
               LABEL_y_dash=None,
               name_of_plot = None):
  y=func(x)
  y_dash = derivative(func, x)
  plt.figure(figsize=(10,8))
  plt.axhline(y=0, color = "blue", linestyle = "--", lw=2)
  plt.axvline(x=0, color = "blue", linestyle = "--", lw=2)
  plt.xlabel(label_x)
  plt.ylabel(label_y)
  plt.title(title)

  if LABEL_y != None and LABEL_y_dash !=None:
    plt.plot(x, y, alpha=ALPHA, label=LABEL_y);
    plt.plot(x, y_dash, alpha=ALPHA, label=LABEL_y_dash);
    plt.legend(fontsize=14)

  else:
    plt.plot(x, y, alpha=ALPHA);
    plt.plot(x, y_dash, alpha=ALPHA);

  plt.savefig(os.path.join(root_plot_dir, f"{name_of_plot}.png"))

# def derivative(f, x, epsilon = 1e-6):
#   return (f(x+epsilon) - (f-epsilon))/(2*epsilon)

def derivative (f , x, delta_x =1e-6):
  return(f(x+delta_x) - f(x))/(delta_x)
```

# Sigmoid Function

The formula for sigmoid is as follows-
$$\sigma (x) = \frac{1}{1+e^{-x}}$$

$$where\ \sigma(x) \in (0, 1),\\
and\ x \in [-\infty, +\infty]$$


```python
def sigmoid(x):
  return tf.keras.activations.sigmoid(x);
```

```python
##python way

def sigmoid(x):
  return 1/(1+np.exp(-x))
```

## Plot of Sigmoid Acticvation fucntion

```python
graph_plot(x, sigmoid, title = "Sigmoid and derivative of Sigmoid function",
           LABEL_y=r"$sigma(x) = \frac{1}{1+e^{-x}}$",
           LABEL_y_dash=r"$\sigma ^\prime(x) = \sigma(x)(1-\sigma(x))$",
           name_of_plot= "sigmoid")
```

# Hyperbolic Tangent Activatoin function



The tanh function formula and curve are as follows

$$tanh(x) = \frac{(e^{x} - e^{-x})}{(e^{x} + e^{-x})}$$

$$where\ \tanh(x) \in (-1, 1),\\
and\ x \in [-\infty, +\infty]$$

```python
def tanh(x):
  return tf.keras.activations.tanh(x)
```

```python
#python way
def tanh(x):
  return np.tanh(x)
```

##Graph plot of hyperbolic tangent fucntion and its derivative 

```python
graph_plot(x, func=tanh, 
           title = "Hyperbolic Tangent Activation Function",
           LABEL_y = r"$tanh(x) = \frac{(e^{x} - e^{-x})}{(e^{x} + e^{-x})}$",
           LABEL_y_dash = r"$tanh^\prime(x) = 1 - tanh^2(x)$",
           name_of_plot = "Hyperbolic Tangent")
```

# Rectified Linear Unit (Relu) Activation Function


ReLU function formula and curve are as follows


$$ReLU(x)= max(x,0)$$

$$where\ ReLU(x) \in (0, x),\\
and\ x \in [-\infty, +\infty]$$

```python
def relu(x):
  return tf.keras.activations.relu(x)
```

##Graph plot of Relu

```python
graph_plot(x, relu,
           title = "ReLU Activation Function",
           LABEL_y = r"$ReLU(x) = max(x,0)$",
           LABEL_y_dash = r"$ReLU \prime (x) = (1,0)$",
           name_of_plot = "ReLU")
```

#Leaky ReLU


Leaky-ReLU function formula and curve are as follows


$$Leaky\_ReLU(x)= \left\{\begin{matrix}
 &x &x\geq0  \\
 &\alpha x &x \lt 0 \\
\end{matrix}\right.$$

$$where x \in [-\infty, +\infty]$$

```python
def leaky_relu(x, alpha =0.5):
  return np.where(x>=0, x, alpha*x)
```

## Graph plot of Leaky ReLU

```python
graph_plot(x, leaky_relu,
           title = "Leaky Relu Activation funciton",
           LABEL_y= r"$Leaky\_ReLU(x)$",
           LABEL_y_dash= r"$Leaky\_ReLU \prime(x)$",
           name_of_plot="LeakyRelu")
           
```

#Exponential Linear Unit (elu)

Leaky-ReLU function formula and curve are as follows


$$elu(x)= \left\{\begin{matrix}
 &x &x\geq0  \\
 &\alpha ⋅ (e^x -1) &x \lt 0 \\
\end{matrix}\right.$$

$$where x \in [-\infty, +\infty]$$

$$ \alpha = scaler\ slope\ of\ negative\ section $$

```python
def elu(x, alpha=1.0):
  return tf.keras.activations.elu(x, alpha)
```

```python
#alternatively

def elu(x, alpha =1.0):
  return np.where(x>=0, x, alpha* (np.exp(x)-1))
```

## Graph plot of elu

```python
graph_plot(x, elu,
           title ="Elu activation function",
           LABEL_y=r"$ elu\ (x)$",
           LABEL_y_dash=r"$elu \prime\ (x)$",
           name_of_plot= "Elu")
```

#PRELu (Parametric Relu)

$$f(y_i) = \left\{\begin{matrix} y_i & y_i>0\\ \alpha_i \cdot y_i & y_i \leq 0 \end{matrix}\right.$$

We look at the formula of PReLU. The parameter α is generally a number between 0 and 1, and it is generally relatively small, such as a few zeros. When α = 0.01, we call PReLU as Leaky Relu , it is regarded as a special case PReLU it.

Above, yᵢ is any input on the ith channel and aᵢ is the negative slope which is a learnable parameter.
* if $\alpha_i=0$, f becomes ReLU
* if $\alpha_i>0$, f becomes leaky ReLU
* if $\alpha_i$ is a learnable parameter, f becomes PReLU
